```python


# Define Rwanda and indicator
country = ["RWA"]
indicators = {
    "NY.GDP.MKTP.KD.ZG": "GDP_Growth_Rate",  # Annual GDP growth %
    "NY.GDP.MKTP.CD": "GDP_Current_USD",       # GDP in current USD
    "NY.GDP.PCAP.CD": "GDP_Per_Capita"         # GDP per capita
}

# Fetch data from 1990 to 2024
start = datetime.datetime(1990, 1, 1)
end = datetime.datetime(2024, 1, 1)

df = wbdata.get_dataframe(indicators, country=country, date=(start, end))

# Clean and reset index
df = df.reset_index()
## df.columns = ['Country', 'Year', 'GDP_Growth_Rate', 'GDP_Current_USD', 'GDP_Per_Capita']
df.columns = ['Year', 'GDP_Growth_Rate', 'GDP_Current_USD', 'GDP_Per_Capita']
df = df.sort_values('Year').reset_index(drop=True)

print(df.head(10))
print(f"\nShape: {df.shape}")
print(f"\nMissing values:\n{df.isnull().sum()}")

# Save to Data folder
df.to_csv("Data/rwanda_gdp_data.csv", index=False)
print("\nData saved to Data/rwanda_gdp_data.csv ✅")
```
